```python
import pandas as pd
import numpy as np
import json
```

```python
ret = []
ert = []
list_key = [
    'article_contentid', 'article_dwelltime', 'article_l1_categories',
    'article_l2_categories', 'article_lan', 'article_publisher',
    'article_title', 'article_update_time', 'dt'
]
with open('tr_data_20170128', 'r') as f:
    for line in f:
        record_dict = {}
        record = json.loads(line)
        ret.append(dict([(key, value) for key, value in record.items() if key in list_key]))
tr = pd.DataFrame(ret)
del ret


with open('tst_article_20170128', 'r') as fi:
    for line in fi:
        record_dict = {}
        record = json.loads(line)
        ert.append(dict([(key, value) for key, value in record.items() if key in list_key]))
tst = pd.DataFrame(ert)
del ert
```

### 筛选出20170124日期之后的文章和它们的category以及weight

```python
#tr_hat = tr[tr.dt > '20170123']
tst.dropna(subset=['article_l1_categories'],how='any',inplace=True)
tst.dropna(subset=['article_l2_categories'],how='any',inplace=True)
tst.dropna(subset=['article_contentid'],how='any',inplace=True)
tr.dropna(subset=['article_l1_categories'],how='any',inplace=True)
tr.dropna(subset=['article_l2_categories'],how='any',inplace=True)
tr.dropna(subset=['article_contentid'],how='any',inplace=True)

c1 = set()
c2 = set()
for i in np.arange(tst.shape[0]):
    for c in tst.iloc[i]['article_l1_categories']:
        c1.add(c['name'])
    for c in tst.iloc[i]['article_l2_categories']:
        c2.add(c['name'])

for i in np.arange(tr.shape[0]):
    for c in tr.iloc[i]['article_l1_categories']:
        c1.add(c['name'])
    for c in tr.iloc[i]['article_l2_categories']:
        c2.add(c['name'])        
```

```python
len(c2)
```

```python
len(c1)
```

```python
#tst.dropna(subset=['article_l1_categories'],how='any',inplace=True)
#tst.dropna(subset=['article_l2_categories'],how='any',inplace=True)
#tst.dropna(subset=['article_title'],how='any',inplace=True)
c1, c2 = list(c1), list(c2)
df1_dict = {}
df2_dict = {}
for title in np.unique(tst['article_contentid']):
    df1_dict[title] = ([title] + [0.0] * len(c1))
    df2_dict[title] = ([title] + [0.0] * len(c2))
for i in np.arange(tst.shape[0]):
    #title = tr.iloc[i]['article_title']
    for c in tst.iloc[i]['article_l1_categories']:
        title = tst.iloc[i]['article_contentid']
        name = c['name']
        weight = c['weight']
        df1_dict[title][c1.index(name)+1] += weight
    for c in tst.iloc[i]['article_l2_categories']:
        title = tst.iloc[i]['article_contentid']
        name = c['name']
        weight = c['weight']
        df2_dict[title][c2.index(name)+1] += weight
```

```python
df1 = pd.DataFrame(df1_dict.values(),columns = ['article_contentid']+c1)
df2 = pd.DataFrame(df2_dict.values(),columns = ['article_contentid']+c2)
df1
```

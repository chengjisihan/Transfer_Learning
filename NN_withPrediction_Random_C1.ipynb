### 根据日期划分训练集和测试集 

```python
data = pd.read_csv('mergeC1_std.csv')
datatrain = data[data.dt < 20170124]
datatest = data[data.dt >20170123]
index = data.columns
x_index = list(index[1:55])
y_index = list(index[56:])
del datatrain['dt']
del datatest['dt']
x_train_data = datatrain.loc[:, x_index]
y_train_data = datatrain.loc[:, y_index]
x_test_data = datatest.loc[:, x_index]
y_test_data = datatest.loc[:, y_index]
x_size = len(x_train_data.iloc[1])
y_size = len(y_train_data.iloc[1])
#use = data.pop('user_id')
#use
#data.insert(0,'user_id',use)
#data
y_train_data

#ind = np.array(y_data.columns)
#ind
```

```python
def add_layer(inputs, in_size, out_size, activation_function=None):
    # add one more layer and return the output of this layer
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
#    print(Weights)
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus_b = tf.matmul(inputs, Weights) + biases
    if activation_function is None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs

# Make up some real data
#x_data = np.linspace(-1,1,300)[:, np.newaxis]
#noise = np.random.normal(0, 0.05, x_data.shape)
#y_data = np.square(x_data) - 0.5 + noise
```

```python
%%time
# define placeholder for inputs to network
xs = tf.placeholder(tf.float32, [None, x_size])
ys = tf.placeholder(tf.float32, [None, y_size])
# add hidden layer
l1 = add_layer(xs, x_size, 60, activation_function=tf.nn.relu)
# add output layer
l2 = add_layer(l1, 60,60, activation_function=tf.nn.relu)
#l3 = add_layer(l2, 60,60, activation_function=tf.nn.relu)
prediction = add_layer(l2, 60, y_size, activation_function=None)

# the error between prediction and real data
loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys - prediction),
                     reduction_indices=[1]))
train_step = tf.train.GradientDescentOptimizer(0.0000005).minimize(loss)

sess = tf.Session()

#for i in range(200):
    # training
#    sess.run(train_step, feed_dict={xs: x_data, ys: y_data})
##    if i % 50 == 0:
        # to see the step improvement
##        print(sess.run(loss, feed_dict={xs: x_data, ys: y_data}))
```

#### 对于每一个用户，计算推荐文章的查准率，并对查准率列表做average得到预测准确率

```python
accuracy = []
recall = []
hit = 0
ad = read[read.dt > 20170123]
#del ad['dt']
#grouped = ad.groupby(ad['user_id'])
#aq = grouped['article_title']
#aq.sum()
```

```python
for index, row in df.iterrows():
    #print
    for col_name in df.columns:
        #print(ad[ad.user_id == index[0]])
        val = row[col_name]
        if (val in np.array(ad[ad.user_id == index[0]].loc[:,['article_contentid']])):
            hit += 1
    length = len(ad[ad.user_id == index[0]].loc[:,['article_contentid']])
    if (length != 0):
        recall.append((hit + 0.0)/length)
    accuracy.append((hit + 0.0)/5)
    hit = 0
    length = 0
m = np.average(accuracy)
r = np.average(recall)
#print(m, r)
#print(recall)
```
